```python
import numpy as np
from ast import literal_eval
import nltk

nltk.download('stopwords')
nltk.download('punkt')
```

```python
features = ['cast','crew','keywords','genres']

for feature in features:
    master_dataset[feature] = master_dataset[feature].apply(literal_eval)
```

```python
def get_director(obj):
    for person in obj:
        if person.get('job') == 'Director':
            return person.get('name')
    return ''

master_dataset['director'] = master_dataset['crew'].apply(get_director)
master_dataset['main_director'] = master_dataset['director'].str.replace(' ', '').str.lower()
```

```python
def get_cast(obj):
    names = []
    for i in range(min(len(obj), 3)):
        names.append(obj[i]['name'])
    return [name.replace(" ", "").lower() for name in names]

master_dataset['cast'] = master_dataset['cast'].apply(get_cast)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def clean_keywords(obj):
    useful = []
    for item in obj:
        word = item['name'].replace(" ", "").lower()
        if len(word) > 1:          # avoid single-letter junk
            useful.append(ps.stem(word))
    return useful

master_dataset['keywords'] = master_dataset['keywords'].apply(clean_keywords)
```

```python
def get_genres(obj):
    return [g['name'].replace(" ", "").lower() for g in obj]

master_dataset['genres'] = master_dataset['genres'].apply(get_genres)
```

```python
def make_soup(row):
    return " ".join(row['keywords']) + " " + \
           " ".join(row['cast']) + " " + \
           row['main_director'] + " " + \
           " ".join(row['genres'])

master_dataset['soup'] = master_dataset.apply(make_soup, axis=1)
```

```python
master_dataset.to_csv("master_dataset_new.csv", index=False)
print("Saved -> master_dataset_new.csv ✔️")
```
