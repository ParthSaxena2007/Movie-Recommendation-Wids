```python
import pandas as pd

df = pd.read_csv("master_dataset_new.csv")
print(df.shape)
df.head()
```

```python
columns_to_keep = ['title','release_date','popularity','main_director','soup']
df = df[columns_to_keep]
print(df.shape)
```

```python
df['popularity'] = pd.to_numeric(df['popularity'], errors='coerce')
df = df.dropna(subset=['popularity'])
df = df.sort_values(by='popularity', ascending=False)
df = df.head(2500)

print(df.shape)
df.head()
```

```python
df = df[['release_date','title','main_director','soup']]
print(df.shape)
df.head()
```

```python
df.to_csv("master_dataset_final.csv", index=False)
print("Saved -> master_dataset_final.csv ✔️")


```
